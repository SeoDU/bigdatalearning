```python
import json
import urllib.request
```

```python
url = "http://openapi.airkorea.or.kr/openapi/services/rest/ArpltnInforInqireSvc/getMsrstnAcctoRltmMesureDnsty"
# 서비스 요청 주소

params = {
    "serviceKey" : "NbR2M9y%2BsYRC2iVB3VToK91EOdm02aeBqTiE%2Fa2E48sNa1LiJqY%2BraDyNd2dz3%2F%2FLkD0I1P%2FdRKbC4ekxjGcug%3D%3D",
    "numOfRows" : 10,
    "pageSize" : 10,
    "pageNo" : 1,
    "startPage" : 1,
    "stationName" : "성북구",
    "dataTerm" : "DAILY",
    "ver" : "1.3",
    "_returnType" : "JSON"
}
# get 방식으로 parameter string을 넘김
# 요청하는 서비스
print(type(params))

params["serviceKey"] = urllib.parse.unquote(params["serviceKey"])
print(type(params))
# urllib에서 자동 변환
params = urllib.parse.urlencode(params)
print(type(params))
# params -> byte type으로 시리얼라이즈
params = params.encode("utf-8")
print(type(params))

req = urllib.request.Request(url, data = params)
# 사람이 넘김
res = urllib.request.urlopen(req)

resStr = res.read()
# 받는 것도 byte
print(type(resStr))
print(resStr)
```

```python
import xml.etree.ElementTree as et
from lxml import etree
```

```python
xmlObj = etree.fromstring(resStr)
xmlRoot = etree.ElementTree(xmlObj).getroot()

etree.dump(xmlRoot)
```

```python
for node in xmlRoot.iter():
    print(node.tag, node.text)
    
itemList = xmlRoot.findall(".//item")

print(len(itemList))
for item in itemList:
    for i in range(len(item)):
        print(item[i].tag, item[i].text)
```
