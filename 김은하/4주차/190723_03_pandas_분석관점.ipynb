## 3. data.describe()

```python
data.describe() # 각 컬럼에 대한 통계값을 보여준다.
```

size, count 간 무슨 차이가 있을까?
size는 미싱 데이터 포함 (null)
count는 미싱 데이터 미포함
평균. 표준편차(std) -> 정규분포
왜 정규분포가 중요? 
큰 수의 법칙과 중심극한정리
중심극한정리 : 수가 크면 정규분포에 가까워진다.
큰 수의 법칙 : 수가 크면 자기만의 고유 분포에 가까워진다.
사실 기계학습하는 사람들은 정규분포에 관심없다.
통계적분석할 때는 mean, std가 정규성 때문에 가장 중요.
정규성을 아는 기본적인 두 가지가 더 있음. 외도(왼쪽으로 쏠렸는가 오른쪽으로 쏠렸는가), 첨도(얼마나 뾰족한가)  

밑에 이 다섯개 boxplot으로 알수 있음 outlier, min max

이 데이터로 1) 정규성 2) boxplot를 알 수 있다.

그런데 숫자로 보는 것보다 그래프로 보는 게 훨씬 편하다.

```python
컬럼이 다섯개였는데 네개만 나옴.
object 데이터는 생략되어서 나온다.
```
