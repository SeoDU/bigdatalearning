```python
class X:
    def __getattr__(self,attr):
        if attr == 'a':
            return '문근영2'
        return '문근영'
```

```python
a = X()
```

```python
a.a
```

```python
a = 1.0
```

```python
getattr(a,'hex')()
```

```python
a.hex()
```

```python
try:
    a = 1/0
except:
    a = 2/0
# error를 발생시킬 때 다르게 실행시켜야 하는데 except 실행시킬 때 또 에러가 나면 또 에러가 난다.
# 즉 try는 중첩되어서 error를 막을 수는 없다.
```
